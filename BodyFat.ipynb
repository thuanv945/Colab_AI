```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import numpy as np

url = 'bodyfat.csv'
data = pd.read_csv(url)
correlation = data.corr()
print(correlation)
sns.heatmap(correlation,annot=True)
data.plot (kind = 'scatter', x = 'Density' , y = 'BodyFat' , figsize = (9,9) , color = 'orange')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Density'] , data['Density']**2 , data['Density']**3, data['Density']**4, data['Density']**5, data['Density']**6, data['Density']**7, data['Density']**8, data['Density']**9, data['Density']**10]).T
model.fit(X,y=data['BodyFat'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['BodyFat']))
range = np.arange(0,2,0.01)
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color = 'green')
```
