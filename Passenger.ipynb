```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import numpy as np

url = 'passenger.csv'
data = pd.read_csv(url)
correlation = data.corr()
print(correlation)
sns.heatmap(correlation,annot=True)
data.plot (kind = 'scatter', x = 'Miles(X)' , y = 'Dollars (Y)' , figsize = (9,9) , color = 'green')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'] , data['Miles(X)']**2 , data['Miles(X)']**3, data['Miles(X)']**4, data['Miles(X)']**5, data['Miles(X)']**6, data['Miles(X)']**7, data['Miles(X)']**8, data['Miles(X)']**9, data['Miles(X)']**10]).T
model.fit(X,y=data['Dollars (Y)'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Dollars (Y)']))
range = np.arange(0,6000,10)
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color = 'yellow')
```
