```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import numpy as np

url = 'USA_Housing.csv'
data = pd.read_csv(url)
correlation = data.corr()
print(correlation)
sns.heatmap(correlation,annot=True)
data.plot (kind = 'scatter', x = 'Price' , y = 'Avg. Area Income' , figsize = (9,9) , color = 'red')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Price'] , data['Price']**2 , data['Price']**3]).T
model.fit(X,y=data['Avg. Area Income'])
print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Avg. Area Income']))
range = np.arange(0,2.5,0.01)
predictor = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color = 'blue')
```
